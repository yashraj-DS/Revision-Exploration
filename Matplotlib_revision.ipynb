#### Bar Chart Styling
| Style Element              | What It Does                        |
| -------------------------- | ----------------------------------- |
| `edgecolor`                | Adds border color around bars       |
| `linewidth`                | Sets border thickness               |
| `hatch`                    | Adds patterns like `///`, `--`, etc |
| `alpha`                    | Controls transparency               |
| `capsize` (for error bars) | Optional  used in stats            |


```python
products = ['Laptop', 'Tablet', 'Mobile', 'Smartwatch']
sales = [150, 120, 300, 80]
x = np.arange(len(products))
plt.bar(x, sales,
        width=0.5,
        color='skyblue',
        edgecolor='black',
        linewidth=1.5,
        hatch='|',         # Hatch Values = '/','\\','*','-','o','+','|'
        alpha=0.8)
plt.xticks(x, products)
plt.xlabel("Product")
plt.ylabel("Units Sold")
plt.title("Product Sales")
plt.show()
```

#### Contour Plot
#### Displays three-dimensional data in two dimensions using contour lines.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

plt.contour(X, Y, Z, cmap='viridis')
plt.title('Contour Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar()
plt.show()
```

| Tool                      | When to Use                                     |
| ------------------------- | ----------------------------------------------- |
| `sharex=True`             | When X-axes are same (e.g., same dates)         |
| `sharey=True`             | When Y-axes should match (e.g., sales in units) |
| `tight_layout()`          | Basic auto-spacing                              |
| `constrained_layout=True` | Smarter layout engine for tight dashboards      |
| `plt.suptitle()`          | Title for all subplots                          |


```python
import matplotlib.pyplot as plt

# Data
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri']
sales_A = [100, 120, 140, 160, 180]
sales_B = [90, 110, 130, 150, 170]
total_sales = [a + b for a, b in zip(sales_A, sales_B)]
regions = ['North', 'South', 'East', 'West']
region_sales = [400, 350, 300, 250]

# Create layout: 2 rows  2 cols
fig, axs = plt.subplots(2, 2, figsize=(12, 6), constrained_layout=True)

# Top Left: Line chart (Daily sales)
axs[0, 0].plot(days, sales_A, label='Store A', marker='o')
axs[0, 0].plot(days, sales_B, label='Store B', marker='s')
axs[0, 0].set_title("Daily Sales Comparison")
axs[0, 0].set_xlabel("Day")
axs[0, 0].set_ylabel("Sales")
axs[0, 0].legend()

# Top Right: Bar chart (Total Sales)
axs[0, 1].bar(days, total_sales, color='skyblue')
axs[0, 1].set_title("Total Sales Per Day")
axs[0, 1].set_xlabel("Day")
axs[0, 1].set_ylabel("Units Sold")

# Bottom: Pie chart (Region sales)
axs[1, 0].pie(region_sales, labels=regions, autopct="%1.1f%%", startangle=90, colors=['orange', 'pink', 'lightgreen', 'lightblue'])
axs[1, 0].set_title("Sales by Region")

# Remove the empty 4th subplot
fig.delaxes(axs[1, 1])

# Add main title
fig.suptitle("Weekly Sales Dashboard", fontsize=16)

plt.show()
```
