```python
# High quality qr generations can cover 30% damage or blocking due to logo inside qrcode
import qrcode
from qrcode.constants import ERROR_CORRECT_H

a = input("Input : ")
file = "C:\\Users\\happy\\OneDrive\\Desktop\\qr_high_quality.png"
qr = qrcode.QRCode(
    version=5,
    error_correction=ERROR_CORRECT_H, 
    box_size=10,
    border=4
)

qr.add_data(a)
qr.make(fit=True)

img = qr.make_image(fill_color="orange", back_color="yellow")
img.save(file)
```

#### Adding logo

```python
import qrcode
from PIL import Image


a = input()
# default parameter with H error correction
qr = qrcode.QRCode(
    error_correction=qrcode.constants.ERROR_CORRECT_H
)
qr.add_data(a)
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white").convert("RGB")

#insert logo in variable and resizing
logo = Image.open("C:\\Users\happy\\OneDrive\\Pictures\\FullLogo2.png")  
logo = logo.resize((60, 60))  
# positing in center: 
pos = ((img.size[0] - logo.size[0]) // 2,
       (img.size[1] - logo.size[1]) // 2)
img.paste(logo, pos)
#save
img.save("C:\\Users\\happy\\OneDrive\\Desktop\\qr_high_quality_logo.png")
```

```python
from PIL import Image
import qrcode
from qrcode.constants import ERROR_CORRECT_H

qr = qrcode.QRCode(error_correction=ERROR_CORRECT_H,box_size=10,border=4)
qr.add_data("https://yashraj-ds.github.io/xxxx/")
qr.make(fit=True)
img = qr.make_image(fill_color="black", back_color="white").convert("RGB")
logo = Image.open("C:\\Users\happy\\OneDrive\\Pictures\\FullLogo_copyfirst.png")
logo_size = 100
logo = logo.resize((logo_size, logo_size))
w, h = img.size
logo_pos = ((w - logo_size) // 2, (h - logo_size) // 2)
img.paste(logo, logo_pos)
img.save("C:\\Users\\happy\\OneDrive\\Desktop\\digital_card2.png")
```

#### Rounded logo with RoundedModuleDrawer&StyledPilImage
#### Easiest way:

```python
import qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers import RoundedModuleDrawer # just learn tough part :)

file = "C:\\Users\\happy\\OneDrive\\Desktop\\qr_rounded.png"
a=input()
qr = qrcode.QRCode(error_correction=qrcode.constants.ERROR_CORRECT_H)
qr.add_data(a)
qr.make()

img = qr.make_image(
    image_factory=StyledPilImage,
    module_drawer=RoundedModuleDrawer()
)

img.save(file)
```

#### Gradient coloring
Easiest way possible

```python
import qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.colormasks import RadialGradiantColorMask
file = "C:\\Users\\happy\\OneDrive\\Desktop\\Gradient_qr.png"
a=input()
qr = qrcode.QRCode()
qr.add_data(a)
qr.make()

img = qr.make_image(
    image_factory=StyledPilImage,
    color_mask=RadialGradiantColorMask(
        center_color=(0, 0, 0),
        edge_color=(0, 200, 255)))

img.save("qr_gradient.png")
```
