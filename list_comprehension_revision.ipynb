##### Q1) From a list of words: ["apple","mango","kiwi","banana","pear"]
Filter only those words whose length is greater than 4.

```python
ls = ["apple","mango","kiwi","banana","pear"]
[x for x in ls if len(x)>4]
```

##### Q2) Given list: [10,15,20,25,30,35]
Filter only numbers greater than 20 and divisible by 5 (use two if conditions).

```python
ls = [10,15,20,25,30,35]
[x for x in ls if x>20 if x%5==0]
```

["a1-b1-c1", "a2-b2-c2"]

using nested loops + join inside comprehension.
